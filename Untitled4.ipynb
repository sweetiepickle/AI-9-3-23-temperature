```python
import numpy as np 
import skfuzzy as fuzz
from skfuzzy import control as ctrl

universe1= np.linspace(18,30,100)
universe2= np.linspace(-5,60,100)
universe3= np.linspace(0,255,256)

setpoint = ctrl.Antecedent(universe1,'Setpoints')
sensor = ctrl.Antecedent(universe2,'Value Sensor')
pwm = ctrl.Consequent(universe3,'PWM')

setpoint['VC']=fuzz.gaussmf(universe1,18,5)
setpoint['C']=fuzz.gaussmf(universe1,20,5)
setpoint['A']=fuzz.gaussmf(universe1,23,5)
setpoint['W']=fuzz.gaussmf(universe1,25,5)
setpoint['H']=fuzz.gaussmf(universe1,28,2)

sensor['VC']=fuzz.gaussmf(universe2,-5,13)
sensor['C']=fuzz.gaussmf(universe2,21,18)
sensor['A']=fuzz.gaussmf(universe2,34,13)
sensor['W']=fuzz.gaussmf(universe2,40,10)
sensor['H']=fuzz.gaussmf(universe2,50,10)

pwm['VL']=fuzz.trapmf(universe3,[0,20,50,60])
pwm['L']=fuzz.trapmf(universe3,[10,25,58,65])
pwm['A']=fuzz.trapmf(universe3,[18,23,180,195])
pwm['H']=fuzz.trapmf(universe3,[170,200,230,240])
pwm['VH']=fuzz.trapmf(universe3,[200,210,250,255])

setpoint.view()
sensor.view()
pwm.view()
     
```

```python
#rules
r1 = ctrl.Rule(antecedent=((setpoint['VC']&sensor['VC']) | (setpoint['C']&sensor['VC']) | (setpoint['VC']&sensor['C'])),consequent=pwm['VL'],label= 'rule 1 ')
r2 = ctrl.Rule(antecedent=((setpoint['A']&sensor['VC']) | (setpoint['W']&sensor['VC']) | (setpoint['C']&sensor['C']) | (setpoint['VC']&sensor['A']) | (setpoint['VC']&sensor['W'])),consequent=pwm['L'],label= 'rule 2 ')
r3 = ctrl.Rule(antecedent=((setpoint['H']&sensor['VC']) | (setpoint['W']&sensor['C']) | (setpoint['A']&sensor['C']) | (setpoint['W']&sensor['A']) | (setpoint['A']&sensor['A'])
| (setpoint['C']&sensor['A']) | (setpoint['A']&sensor['W']) | (setpoint['C']&sensor['W']) | (setpoint['VC']&sensor['H'])),consequent=pwm['A'],label= 'rule 3 ')
r4 = ctrl.Rule(antecedent=((setpoint['H']&sensor['C']) | (setpoint['H']&sensor['A']) | (setpoint['W']&sensor['W']) | (setpoint['A']&sensor['H']) | (setpoint['C']&sensor['H'])),consequent=pwm['H'],label= 'rule 4 ')
r5 = ctrl.Rule(antecedent=((setpoint['H']&sensor['H']) | (setpoint['H']&sensor['W']) | (setpoint['W']&sensor['H'])),consequent=pwm['VH'],label= 'rule 5 ')

system=ctrl.ControlSystem(rules=[r1,r2,r3,r4,r5])
sim = ctrl.ControlSystemSimulation(system,flush_after_run= 100*100+1)
upsampled1 = np.linspace(18,30,100)
upsampled2 = np.linspace(-5,60,100)
upsampled3 = np.linspace(0,255,256)

x,y = np.meshgrid(upsampled1,upsampled2)
z= np.zeros_like(x)
for i in range(100):
  for j in range(100):
    sim.input['Setpoints']= x[i,j]
    sim.input['Value Sensor']=y[i,j]
    sim.compute()
    z[i,j]=sim.output['PWM']
```
